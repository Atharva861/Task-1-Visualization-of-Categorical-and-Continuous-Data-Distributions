```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_26346.csv', skiprows=4)
print(df.head())
```

```python
np.random.seed(42)
ages = np.random.normal(loc=35, scale=15, size=1000) 
ages = ages[(ages > 0) & (ages < 100)]
```

```python
df = pd.DataFrame({'Age': ages})
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age in Sample Population')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.grid(True)
plt.show()
```

```python
gender_data = np.random.choice(['Male', 'Female', 'Other'], size=1000, p=[0.48, 0.50, 0.02])
df_gender = pd.DataFrame({'Gender': gender_data})
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=df_gender, x='Gender', palette='pastel')
plt.title('Gender Distribution in Sample Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True, axis='y')
plt.show()
```

```python
n = 1000
data = {
    'Age': np.random.normal(35, 12, n).clip(0, 90),  # ages between 0 and 90
    'Gender': np.random.choice(['Male', 'Female', 'Other'], size=n, p=[0.48, 0.50, 0.02]),
    'Education': np.random.choice(['Primary', 'Secondary', 'Graduate', 'Post-Graduate'], size=n, p=[0.2, 0.4, 0.3, 0.1])
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Education', data=df, palette='muted')
plt.title('Bar Chart of Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.grid(True, axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Gender', bins=20, element='step', stat='count', common_norm=False)
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Education', hue='Gender', palette='pastel')
plt.title('Education Level by Gender')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.grid(True, axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df[df['Education'] == 'Graduate'], x='Age', bins=15, kde=False, color='orange')
plt.title('Age Distribution for Graduates')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
